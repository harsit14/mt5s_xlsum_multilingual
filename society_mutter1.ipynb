```python
# Creating new path for the downloaded dataset
dummy_data_path='/content/xlsum/dummy'
# Importing necessary python libraries
import pandas as pd
import os
# Initializing train data and test data variables
train_data = os.listdir(dummy_data_path)
train_df = pd.DataFrame(train_data)
# Printing train set dataframe
train_df
```

```python
# Defining the parameters of our datasets to be used
max_input_length = 2048
max_target_length = 128

# Tokenizing model inputs, processesing each input into required length
def preprocess_function(examples):
    model_inputs = tokenizer(
        examples["text"], max_length=max_input_length, truncation=True
    )
    # Set up the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            examples["summary"], max_length=max_target_length, truncation=True
        )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
# storing tokenized datset after preprocess function applied on the dataset
tokenized_datasets = language_dataset.map(preprocess_function, batched=True)
```

```python
# Defining a strong abseline for our summarization model, ie. three sentence summary
from nltk.tokenize import sent_tokenize


def three_sentence_summary(text):
    return "\n".join(sent_tokenize(text)[:3])


print(three_sentence_summary(language_dataset["train"][2]["text"]))
```

```python
# Defining evaluation baseline for our dataset to be used in rogue analysis
def evaluate_baseline(dataset, metric):
    summaries = [three_sentence_summary(text) for text in dataset["text"]]
    return metric.compute(predictions=summaries, references=dataset["summary"])
```

```python
import pandas as pd
# Printing rouge metric evaluation
# Evaluating on validation set
score = evaluate_baseline(language_dataset["validation"], rouge_score)
rouge_names = ["rouge1", "rouge2", "rougeL", "rougeLsum"]
rouge_dict = dict((rn, round(score[rn].mid.fmeasure * 100, 2)) for rn in rouge_names)
rouge_dict
```

```python
# Definiting arguments to be used in the training as parameter
from transformers import Seq2SeqTrainingArguments

batch_size = 1
num_train_epochs = 5
# Show the training loss with every epoch
# logging_steps = len(tokenized_datasets["train"]) // batch_size
model_name = model_checkpoint.split("/")[-1]

args = Seq2SeqTrainingArguments(
    output_dir=f"{model_name}-finetuned-multilingual-xlsum-new",
    evaluation_strategy="epoch",
    learning_rate=5.6e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=num_train_epochs,
    predict_with_generate=True,
    logging_steps=10,
    gradient_accumulation_steps=8,
    push_to_hub=True,
)
```

```python
import numpy as np

# Defining metrics to be computed during the training loop
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    # Decode generated summaries into text
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    # Replace -100 in the labels as we can't decode them
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    # Decode reference summaries into text
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)
    # ROUGE expects a newline after each sentence
    decoded_preds = ["\n".join(sent_tokenize(pred.strip())) for pred in decoded_preds]
    decoded_labels = ["\n".join(sent_tokenize(label.strip())) for label in decoded_labels]
    # Compute ROUGE scores
    result = rouge_score.compute(
        predictions=decoded_preds, references=decoded_labels, use_stemmer=True
    )
    # Extract the median scores
    result = {key: value.mid.fmeasure * 100 for key, value in result.items()}
    return {k: round(v, 4) for k, v in result.items()}
```

```python
# Loading datacollator to pad the input and label with appropriate padding as required by the training loop
from transformers import DataCollatorForSeq2Seq

data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
```

```python
from transformers import Seq2SeqTrainer
# Defining training parameters for trainer
trainer = Seq2SeqTrainer(
    model,
    args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
# Clearing GPU Cache before running the training loop
import torch
torch.cuda.empty_cache()
```

```python
# Training starts here
trainer.train()
```

```python
# Pushing the trained model to the huggingface Hub
trainer.push_to_hub(commit_message="Training complete", tags="multilingual model")
```
